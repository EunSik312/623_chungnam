```
import cv2
import numpy as np
from matplotlib import pyplot as plt
from ipywidgets import interact, IntSlider

def show_hsv_color(h=0, s=255, v=255):
    # HSV 색상을 BGR로 변환 (OpenCV는 BGR 사용)
    hsv_color = np.uint8([[[h, s, v]]])
    bgr_color = cv2.cvtColor(hsv_color, cv2.COLOR_HSV2BGR)

    # 시각화용 이미지 생성 (100x100 픽셀)
    img = np.full((100, 100, 3), bgr_color[0][0], dtype=np.uint8)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title(f'H: {h}, S: {s}, V: {v}')
    plt.axis('off')
    plt.show()

# 슬라이더 UI
interact(show_hsv_color,
         h=IntSlider(min=0, max=180, step=1, value=0, description='Hue'),
         s=IntSlider(min=0, max=255, step=1, value=255, description='Saturation'),
         v=IntSlider(min=0, max=255, step=1, value=255, description='Value'))
```
